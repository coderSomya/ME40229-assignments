```python
#constants

max_iter = 1000
tolerance = 1e-6
learning_rate = 0.1
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x,y):
    return x**2 + y**2 - x*y

def grad_f(x,y):
    return (2*x-y,2*y-x)
```

```python
x,y = 1,1

points = []
points.append((x,y))
```

```python
steps = 0
converged = False

while(steps<max_iter):
    steps+=1
    grad_x,grad_y = grad_f(x,y)
    
    x_new = x-learning_rate*grad_x
    y_new = y-learning_rate*grad_y
    
    dist = np.linalg.norm([x-x_new, y-y_new])
    
    if(dist<tolerance):
        converged=True
        break
        
    x = x_new
    y = y_new
    
    points.append((x,y))
```

```python
print(converged)
```

```python
iterations = [i for i in range(len(points))]
values = [f(x,y) for (x,y) in points]
```

```python
plt.plot(iterations, values, 'r-o', markersize=5, label='Steepest Descent Path', color='blue')


# Add labels and title
plt.xlabel('Iterations')
plt.ylabel('F(x,y)')
plt.title('Gradient Descent')

# Show the plot
plt.grid(True)
plt.savefig('gradient_descent.png',bbox_inches='tight',facecolor='white')
plt.show()
```
