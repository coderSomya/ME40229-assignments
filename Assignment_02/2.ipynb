```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#constants

t_0 = 80
t_inf = 20

rho = 1000
heat_capacity = 2000
surface_area = 0.5
volume = 0.25
timestamp = [t for t in range(121)]

hs = [5,10,15]
```

```python
line_styles = ['-', '--', '-.']
markers = ['o', 's', 'D']
colors = ['blue', 'green', 'red']
```

```python
plt.figure(figsize=(10, 6))

for i,h in enumerate(hs):
    temp_t = []
    factor = -1 * h * surface_area / (rho * volume * heat_capacity)
    
    for t in timestamp:
        temp = t_inf + (t_0 - t_inf) * np.exp(t* factor)
        temp_t.append(temp)
    
    plt.plot(timestamp, temp_t, linestyle=line_styles[i], color=colors[i], 
             marker=markers[i], label=f'h = {h}', markevery=25)

plt.xlabel('Time (s)')
plt.ylabel('Temperature (°C)')

plt.title('Temperature vs Time for Different Heat Transfer Coefficients')

plt.legend()

plt.grid(True)
plt.savefig("temperature_gradient.png")
plt.show()
```
