```python
# Fourier Transform
```

```python
import numpy as np
A = 10
x = np.arange(0,128)
y = np.sin(2*np.pi*x/128) + 0.5*np.sin(10*2*np.pi*x/128)
```

```python
import matplotlib.pyplot as plt
plt.plot(x,y)
```

```python
import scipy.fft as sft
```

```python
transform = sft.fft(y)
```

```python
sp = sft.fftshift(transform)
```

```python
freq = sft.fftshift(sft.fftfreq(x.shape[-1]))
```

```python
plt.plot(freq, sp.real, freq, sp.imag)
```
