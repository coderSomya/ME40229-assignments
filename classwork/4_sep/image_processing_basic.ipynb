```python
import matplotlib.pyplot as plt
import numpy as np

redball = cv2.imread('red_ball.jpg', cv2.IMREAD_COLOR)
redball = cv2.cvtColor(redball,cv2.COLOR_BGR2RGB)
greenball = cv2.imread('green_ball.jpg', cv2.IMREAD_COLOR)
greenball = cv2.cvtColor(greenball,cv2.COLOR_BGR2RGB)

redball_1 = cv2.erode(redball, np.ones((3,3), np.uint8), iterations=1)
plt.imshow(redball_1)
```

```python
erode = cv2.imread('erode.png', cv2.IMREAD_COLOR)
erode = cv2.dilate(erode, np.ones((3,3), np.uint8), iterations=2)
plt.imshow(erode)
```

```python
## image_opening = erode + dilate
## image_closing = dilate + erode

opened = cv2.morphologyEx(erode, cv2.MORPH_OPEN, np.ones((3,3), np.uint8))
plt.imshow(opened)
```
