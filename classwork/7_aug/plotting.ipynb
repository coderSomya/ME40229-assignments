Matplotlib and Seaborn

```python
def f(x):
    return -x**2 + 4*x
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-4, 4)
y = f(x)
```

```python
plt.plot(x,y, x, y**2)
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
def f_prime(x):
    return -2*x+4
```

```python
x_p = [-1,0,1,2]
for x0 in x_p:
    slope = f_prime(x0)
    y0 = f(x0)
    

    # y = mx + c
    c = y0 - slope*x0
    delta = 0.01

    def line(x):
        return slope*x + c
    

    plt.plot([x0-delta, line(x0-delta)], [x0+delta, line(x0+delta)], '-r')
```

```python
# CYCLING ACROSS A MARKER ARRAY

markers = ['x', 's', '^', 'o', 'v']

from itertools import cycle
marker_cycle = cycle(markers)


x = np.linspace(0,10,10)
y_a = [x,x**2, np.sqrt(x), np.log(x+1)]

plt.figure(figsize=(6,4))

for y in y_a:
    plt.plot(x,y, marker=next(marker_cycle))
```

```python
w = 1.5
A = 10
phi = 0
N = 30

x1 = np.linspace(0,2.5*np.pi/w,30)


displacement = [A*np.sin(w*t+phi) for t in x1]
velocity = [A*w*np.cos(w*t+phi) for t in x1]
acceleration = [-A*w*w*np.sin(w*t+phi) for t in x1]
```

```python
plt.plot(x1,displacement)
```

```python
plt.plot(displacement, velocity)
```

```python
## subplots
plt.figure(figsize=(10,6))
plt.title("Simple Harmonic Motion")
plt.subplot(3,2,1)
plt.plot(x1,displacement)
plt.ylabel('displacement')

plt.subplot(3,2,2)
plt.plot(x1,velocity)
plt.ylabel('velocity')


plt.subplot(3,1,3)
plt.plot(x1,acceleration)
plt.ylabel('acceleration')


plt.xlabel('time')
plt.show()
```

```python
## dissipative system

decay_rate = 0.3

y_prime = [A*np.exp(-decay_rate*t)*np.sin(w*t+phi) for t in x1]
plt.plot(x1, y_prime)
```
