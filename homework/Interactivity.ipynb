```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interactive
```

```python
def plot_sin(k=1):
    
    x = np.linspace(0,2*k*np.pi)
    y = np.sin(x)
    
    plt.plot(x,y)
```

```python
plot_sin(2)
```

```python
w = interactive(plot_sin, k = (0.1,10,0.05))
w
```
