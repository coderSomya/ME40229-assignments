```python
import numpy as np
import matplotlib.pyplot as plt
```

$\exp(x) - 3x^2=0$

```python
x = np.linspace(-2,5)
y = np.exp(x) - 3*(x**2)
plt.plot(x,y)
plt.axhline(y=0, color='r', linestyle='-')
plt.ylim(-2,5)
```

```python
def f(x):
    return np.log(3)+2*np.log(x)
```

```python
def perform_iterations(initital_guess=0):
    n_iter = 50

    x_vals =np.zeros(n_iter)
    x_vals[0] = initial_guess
    
    for i in np.arange(1,n_iter):
        x_vals[i]=f(x_vals[i-1])
    
    return x_vals
```

```python
initial_guess = 0.99
x = perform_iterations(initial_guess)
print(x[-1])
```

```python
guesses = np.linspace(1,5,200)

for guess in guesses:
    x = perform_iterations(guess)
    y = f(x)
    plt.plot(np.arange(0,n_iter), x)
```
